```python
from sklearn.cluster import KMeans
```

```python
trydata=seprate[['Average Cost for two','Price range','Aggregate rating','Votes','Rating_no']]
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,init='k-means++', 
           n_init=10,
         max_iter=300, tol=1e-04, random_state=0,)
    km.fit(seprate[['Average Cost for two','Price range','Aggregate rating','Votes','Rating_no']])
    distortions.append(km.inertia_)
```

```python
km4= KMeans(n_clusters=3,init='k-means++', 
           n_init=10,
         max_iter=300, tol=1e-04, random_state=0,)
```

```python
predict=km4.predict(trydata)
```

```python
km4.fit(trydata)
```

```python
plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
